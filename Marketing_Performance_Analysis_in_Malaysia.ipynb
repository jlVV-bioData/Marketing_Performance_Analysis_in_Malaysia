##### Deducible values from other columns

We realise there are missing values that are deduble doing formulas and operations thorugh values of ohter variables.

    Columns 'Start_Date', 'End_Date' and 'Duration'

We found that sometimes the end date is earlier than the start date. This also affects the duration values. Therefore we corrected this

```python
# ---{CORRECTION OF COLUMNS'Start_Date', 'End_Date' and 'Duration'}----

# [Determination of missing dates and duration]

# Convert the date columns to datetime format
df1['Start_Date'] = pd.to_datetime(df1['Start_Date'], format='%d/%m/%Y', errors='coerce')
df1['End_Date'] = pd.to_datetime(df1['End_Date'], format='%d/%m/%Y', errors='coerce')

# Swap values of Start_Date and End_Date where End_Date is earlier than Start_Date
mask = df1['End_Date'] < df1['Start_Date']
df1.loc[mask, ['Start_Date', 'End_Date']] = df1.loc[mask, ['End_Date', 'Start_Date']].values

# Calculate missing values in Duration
df1['Duration'] = df1['Duration'].fillna((df1['End_Date'] - df1['Start_Date']).dt.days)

# Ensure there are no negative values in Duration
df1['Duration'] = df1['Duration'].apply(lambda x: x if x >= 0 else None)

# Fill missing values in Start_Date and End_Date using Duration
df1['Start_Date'] = df1.apply(
    lambda row: row['End_Date'] - pd.Timedelta(days=row['Duration']) if pd.isnull(row['Start_Date']) and pd.notnull(row['End_Date']) and pd.notnull(row['Duration']) else row['Start_Date'],
    axis=1
)
df1['End_Date'] = df1.apply(
    lambda row: row['Start_Date'] + pd.Timedelta(days=row['Duration']) if pd.isnull(row['End_Date']) and pd.notnull(row['Start_Date']) and pd.notnull(row['Duration']) else row['End_Date'],
    axis=1
)

# Verify the changes
print(df1[['Start_Date', 'End_Date', 'Duration']])

# Check reduction of missing values in columns 'Start_Date', 'End_Date', 'Duration'
valores_nulos_df1 = df1.isnull().sum()
print(valores_nulos_df1)

# Heatmap of missing values with corrected Start_Date, End_Date, and Duration
plt.figure(figsize=(12, 8))
sns.heatmap(df1.isnull(), cbar=False, cmap='viridis')
plt.title("Correction of missing values in ['Start_Date'], ['End_Date'] and ['Duration']")
plt.show()
```

    Columns 'ROI', 'Revenue', 'Spend' and 'CTR'

```python
#----{CORRECTION OF COLUMNS 'ROI', 'Revenue', 'Spend' and 'CTR' (whose values are deducible from the data of other columns through formulas and mathematical operations)}----

# Calculate missing values for columns 'CTR' and 'ROI'
df1['CTR'] = df1['CTR'].fillna(df1['Clicks'] / df1['Impressions'])
df1['ROI'] = df1['ROI'].fillna(df1['Revenue'] / df1['Spend'])

# Calculate missing values for ROI
df1['ROI'] = df1.apply(
    lambda row: (row['Revenue'] / row['Spend']) * 100 if pd.isnull(row['ROI']) and pd.notnull(row['Revenue']) and pd.notnull(row['Spend']) and row['Spend'] != 0 else row['ROI'],
    axis=1
)

# Calculate missing values for Revenue
df1['Revenue'] = df1.apply(
    lambda row: (row['ROI'] * row['Spend']) / 100 if pd.isnull(row['Revenue']) and pd.notnull(row['ROI']) and pd.notnull(row['Spend']) and row['Spend'] != 0 else row['Revenue'],
    axis=1
)

# Calculate missing values for Spend
df1['Spend'] = df1.apply(
    lambda row: row['Revenue'] / (row['ROI'] / 100) if pd.isnull(row['Spend']) and pd.notnull(row['Revenue']) and pd.notnull(row['ROI']) and row['ROI'] != 0 else row['Spend'],
    axis=1
)

# Check reduction of missing values
valores_nulos_df1 = df1.isnull().sum()  # Show missing values for each column
print(valores_nulos_df1)

# Create a heatmap of the missing values
plt.figure(figsize=(12, 8))
sns.heatmap(df1.isnull(), cbar=False, cmap='viridis')
plt.title("Correction of missing values in ['ROI'], ['Revenue'], and ['Spend']")
plt.show()

# Show the size of df1
df1.shape
```

##### Non deducible values

Is seen missing values are randomly distributed on the dataset, so is complicated to fix them one by one doing a KNN predictor for each column to get very accurate results values. Is decided to make a general KNN prediction of numeric 

```python
#---{CORRECTION OF MISSING VALUES IN NUMERIC COLUMS USING KNN}---

# [Data Scaling]

# Check distances between max and min values in each column
columns_of_interest = ['Budget', 'Impressions', 'Clicks', 'CTR', 'Conversions', 'CR', 'Revenue', 'ROI', 'Duration', 'Spend']  # Select columns of interest
max_values = df1_nulos2[columns_of_interest].max()  # Calculate max and min values
min_values = df1_nulos2[columns_of_interest].min()

# Create a DataFrame with max and min values to better visualize and choose
extreme_values = pd.DataFrame({
    'Max': max_values,
    'Min': min_values
})
# Display the DataFrame
print(extreme_values)

# Import the library for scaling
from sklearn.preprocessing import StandardScaler 

# Select the columns that may be related to the 'Conversions' column to be scaled
columns_to_scale = ['Spend', 'Revenue', 'Duration', 'Impressions', 'Clicks', 'Conversions', 'Budget']

# Create a copy of the DataFrame to avoid modifying the original
df1_scaled = df1_nulos2.copy()

# Apply the scaler
scaler = StandardScaler()
df1_scaled[columns_to_scale] = scaler.fit_transform(df1_scaled[columns_to_scale])

df1_scaled[columns_to_scale]
```

```python
# --{FILL MISSING NUMERIC VALUES WITH "KNN IMPUTER"}--

# Import the KNN Imputer library
from sklearn.impute import KNNImputer

# Select the columns to be used for imputation
columns_to_impute = ['Conversions', 'Impressions', 'Clicks', 'CR', 'Budget', 'Customer_Satisfaction', 'Campaign_Effectiveness', 'Spend']

# Apply KNN Imputer
imputer = KNNImputer(n_neighbors=5)
df1_scaled[columns_to_impute] = imputer.fit_transform(df1_scaled[columns_to_impute])

# Verify the changes
print("Missing values after imputation:")
print(df1_scaled['Conversions'].isnull().sum())

print("First rows of the DataFrame after imputation:")
df1_scaled

missing_values_count = df1_scaled.isnull().sum()
print(missing_values_count)

plt.figure(figsize=(12, 8))
sns.heatmap(df1_scaled.isnull(), cbar=False, cmap='viridis')
plt.title('Correction of missing numeric values with KNN Imputer')
plt.show()
```

```python
# [FILL MISSING VALUES OF NON-NUMERIC VARIABLES COLUMNS using statistical operations]
# Fill missing values
for column in ['Channel', 'Product_Category', 'Region', 'Market_Segment']:
    mode = df1_scaled[column].mode()[0]
    df1_scaled[column].fillna(mode, inplace=True)

# Verify the changes
print(df1_scaled[['Channel', 'Product_Category', 'Region', 'Market_Segment']].isnull().sum())

plt.figure(figsize=(12, 8))
sns.heatmap(df1_scaled.isnull(), cbar=False, cmap='viridis')
plt.title('Missing values filled with mode')
plt.show()
```

```python
# Descale the specified columns
columns_to_descale = ['Spend', 'Revenue', 'Duration', 'Impressions', 'Clicks', 'Conversions', 'Budget']
df1_descaled = df1_scaled.copy()
df1_descaled[columns_to_descale] = scaler.inverse_transform(df1_scaled[columns_to_descale])

# Verify the changes
print(df1_descaled[columns_to_descale].head())

df1_descaled
```

- Realiza anÃ¡lisis de clusters o segmentaciÃ³n si es relevante.

```python
# Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select the numeric columns of interest
numeric_columns = ['Budget', 'Spend', 'Impressions', 'Clicks', 'CTR', 'Conversions', 'CR', 'Revenue', 'ROI', 'Duration']

# Scale the data
scaler = StandardScaler()
df_final_scaled = scaler.fit_transform(df_final[numeric_columns])

# Determine the optimal number of clusters using the elbow method
sse = []
for k in range(1, 8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_final_scaled)
    sse.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 8), sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.title('Elbow Method to Determine the Optimal Number of Clusters')
plt.show()

# Based on the elbow plot, select the optimal number of clusters
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_final['Cluster'] = kmeans.fit_predict(df_final_scaled)

# Cluster analysis
for i in range(n_clusters):
    print(f"Cluster {i}:")
    print(df_final[df_final['Cluster'] == i][numeric_columns].describe())
    print("\n")

# Cluster visualization
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_final, x='Budget', y='Revenue', hue='Cluster', palette='viridis')
plt.title('Marketing Campaign Clusters')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.legend(title='Cluster')
plt.show()

# Analysis of the characteristics of each cluster
for i in range(n_clusters):
    print(f"Cluster Analysis {i}:")
    print(f"1. Observe the main characteristics of the campaigns in this cluster.")
    print(f"2. Identify common patterns and trends among the campaigns.")
    print(f"3. Compare key metrics such as Budget, Revenue, ROI, etc.")
    print(f"4. Use this information to design specific marketing strategies for each group.")
    print(f"Conclusion: Cluster {i} shows that...")
    print(f"As a professional analyst, I have observed that the campaigns in this cluster have similar characteristics in terms of budget and performance. This suggests that we can apply specific marketing strategies to optimize the performance of these campaigns.")
    print("\n")
```
